#### Myntra Scraping

```python
import os
import requests
import time
from selenium import webdriver
from bs4 import BeautifulSoup as soup
from selenium.webdriver.common.by import By
```

```python
product = "Watches"
# Define the path to Chrome WebDriver executable
driver = webdriver.Chrome()

# Navigate to URL
driver.get("https://www.myntra.com/kurta?rawQuery=" + product)

# Wait for the page to load (you can adjust time as per the page load)
time.sleep(5)

# Extract the page source after the page loads
myntra_text = driver.page_source

# Parse the page source with BeautifulSoup
myntra_html = soup(myntra_text, "html.parser")
```

```python
# Find product titles and prices
product_titles = myntra_html.find_all("h4", class_="product-product")  # Adjust the class name based on actual HTML
product_prices = myntra_html.find_all("span", class_="product-discountedPrice")  # Adjust as needed

# Loop through each product and print the title and price
for title, price in zip(product_titles, product_prices):
    product_name = title.get_text(strip=True)
    product_price = price.get_text(strip=True)
    print(f"Product Name: {product_name}, Price: {product_price}")
```

```python
# Find product details such as color, size, material
product_details = myntra_html.find_all("div", class_="product-productMetaInfo")
for i in product_details:
    print(i.get_text())
```

```python
# Find image URLs for products
product_images = myntra_html.find_all("div", class_="product-sliderContainer")
product_images
```

```python
# Find product ratings and reviews (this might vary depending on the structure of the page)
product_ratings = myntra_html.find_all("div", class_="product-ratingsContainer")
product_reviews = myntra_html.find_all("div", class_="product-ratingsCount")

for rating, reviews in zip(product_ratings, product_reviews):
    product_rating = rating.get_text(strip=True)
    review_count = reviews.get_text(strip=True)
    print(f"Rating: {product_rating}, Reviews: {review_count}")
```
